# Observation: all the columns are of type int. age,op_year,aux_nodes are independent features. surv stat is class variable.

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

#Observation: there are no NaN values in any column

```python
df['surv_stat'].unique()
```

#Observation: two diff classes in the column surv_stat

```python
df['surv_stat'].value_counts()
```

```python
sns.countplot(x='surv_stat',data=df)
plt.tight_layout()
plt.title('count plot of surv_stat')
plt.show()
```

#Observation: Not a balanced set

```python
plt.close()
sns.FacetGrid(data=df,hue='surv_stat',height=5).map(sns.histplot,'age',kde=True,bins=10).add_legend()
plt.tight_layout()
plt.title('histogram of age')
plt.show()
```

#Observation:
*   Most of the survivors age lies in between 48 to 58
*   Most of the nonsurvivors age lies in between 45 to 55

```python
sns.FacetGrid(data=df,hue='surv_stat',height=5).map(sns.histplot,'aux_nodes',bins=10).add_legend()
```

#observation:


*   more survivors are found with aux_nodes from 0-5
*   but there are nearly 50 members that did not survive with the same    aux_nodes(from 0-5)




```python
sns.FacetGrid(data=df,hue='surv_stat',height=5).map(sns.histplot,'op_year',bins=10).add_legend()
```

Observation:  

*   Operation year does not play an important role in predicting whether a person survives or not. I have come to this point based on common sense.
*   So, the only imp features are age and aux_nodes




```python
plt.close()
sns.pairplot(data=df,hue='surv_stat',palette='copper')
plt.title('pairplot for habermans dataset')
plt.tight_layout()
plt.show()
```

#Observation: Useful features are age and aux_nodes

```python
print('Mean: ',df.groupby('surv_stat')['age'].mean())
print('\n')
print('Median: ',df.groupby('surv_stat')['age'].median())
```

#Observation: Mean and median values of age are almost equal. Hence there are no outliers.

```python
print('Mean: ',df.groupby('surv_stat')['aux_nodes'].mean())
print('\n')
print('Median: ',df.groupby('surv_stat')['aux_nodes'].median())
```

#Observation: Mean and median values of aux_nodes have huge difference. Hence there are some outliers.

```python
plt.close()
sns.boxplot(x='surv_stat',y='age',data=df,hue='surv_stat')
plt.title('boxplot for surv_stat and age')
plt.tight_layout()
plt.show()
```

#Observation: As mentioned earlier, there are no outliers

```python
plt.close()
sns.boxplot(x='surv_stat',y='aux_nodes',data=df,hue='surv_stat')
plt.title('boxplot for aux_nodes and age')
plt.tight_layout()
plt.show()
```

#Observation: As mentioned earlier, there are some outliers

```python
plt.close()
sns.set_style(style='whitegrid')
sns.scatterplot(y='aux_nodes',x='age',data=df,hue='surv_stat',palette='flare')
plt.title('scatterplot for aux_nodes and age')
plt.tight_layout()
plt.show()
```

#Observations: 
*   30<age<40 and aux_nodes<10 have high chances of survival
*   aux_nodes=0 have higher chances of surviving
*   40<age<60 and aux_nodes <=2 have higher chances of surviving
*   40<age<60 and aux_nodes <=2 have higher chances of surviving
*   60<age<74 and aux_nodes <=1 have higher chances of surviving
*   people with 45<age<58 and aux_nodes >7 have lesser chances of surviving
*   people with age>76 have less chances of survival
